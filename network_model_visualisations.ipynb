```python
import plotly
import plotly.offline as py
import plotly.graph_objs as go
import pandas as pd
```

```python
plotly.__version__
```

## Using BaseSympModel

```python
track_states_df = pd.read_csv('BaseSympModel0_99.csv')
track_states_df.head()
```

```python
disease_states = ['Exposed', 'Infected_Presymptomatic', 'Infected_Symptomatic', 'Infected_Asymptomatic', 'Hospitalized', 'Recovered', 'Fatalities']
disease_state_tracker_plot = go.Figure()

for state in disease_states:
    disease_state_tracker_plot.add_trace(go.Scatter(x=track_states_df['Time'],
                                                    y=track_states_df[state], 
                                                    name=state,
                                                    mode='lines+markers',
                                                    hovertemplate='%{y}'))
disease_state_tracker_plot.layout.xaxis.title = 'Day'
disease_state_tracker_plot.layout.yaxis.title = 'Population'
disease_state_tracker_plot.layout.hovermode = 'x'
disease_state_tracker_plot.layout.title = 'Disease States Progress'
disease_state_tracker_plot.show()
```
